# Task 1
**Password Validaiton**

Write a program that asks the user to enter a password and checks whether the password is valid according to the following rules. If all conditions are met, print "Valid password". Otherwise, print "Invalid password".

```python
#Define two lists of the conditions special character and digits.
s_character = "!@#$%^&*()-_=+[]{}|;:'\",.<>?"
digit = "1234567890"
```

```python
#Define a function valid_password(p) and give the conditional statements less than 6 or greater than 12 = False.
def valid_password(p):
    if len(p) < 6 or len(p) > 12:
        return False

#Then, define the conditions as = 0 in order to get True or False statements later on.
    l_special = 0
    l_upper = 0
    l_digit = 0

#Loop statement for l in p.
    for l in p:
        if l in digit:
            l_digit += 1
        elif l.isupper():
            l_upper += 1
        elif l in s_character:
            l_special += 1

#If the conditions are met, the statement will return True.
    if l_upper and l_digit and not l_special:
        return True
    return False

#Here, you fill in the password and get the two print-outcomes whether or not the conditions are met.
while True:
    p = input("Enter  your password: ")

    if valid_password(p):
        print("Valid password.")
        break
    else:
        print("Invalid Password. Must contain 6-12 characters, one uppercase letter, one digit and without any special character.")
```

**Data Dispersion Measurement**

Given a list of numbers L = [x1, x2, ..., xk], we can measure the dispersion level between these numbers using a metric called d, calculated as follows:

    d = (Σk i=1 [xi-m])/k
    
where k is the number om elements on the list L and m is the average of these elements in L.

Write a function that takes a list as a parameter and returns the value of metric d. Round the number to two decimal places.

Test your function with

(1) list_1 = [4, 1, 4, 1, 3, 5]

(2) list_2 = [43, 31, 41, 40, 40, 48, 29, 48, 23, 33]

```python
#Define function
def math(list_):
    sum_xim = 0

#Define the average number 
    avg = sum(list_) / len(list_)
    
#Define sum_xim for xi value in list_.
    for xi in list_:
        sum_xim += abs(xi - avg)
        
#Return the rounded sum of sum_xim divided by the number of elements in list_ with 2 decimals.
    return round(sum_xim/len(list_), 2)
```

```python
#Call function
list_1 = [4,1,4,1,3,5]
result = math(list_1)

print(result)
```

```python
#Call function
list_2 = [43, 31, 41, 40, 40, 48, 29, 48, 23, 33]
answer = math(list_2)

print(answer)
```

**4.1 Bus Stop Information**

The following map shows the stops of bus line A in the city, including the stop ID i (i = 1, 2, ..., 7) and its coordinates (xi, yi). For example, the Stop 1 has the coordinates (2, 1), and Stop 7 has the coordinates (7, 6). (One unit is one kilometer.)

Store the bus stop information in a dictionary using the stop ID as the key and the coordinates as the value.

```python
#Define a dictionary with the stop ID and coordinates.
stop_information = {"1":[2,1],"2":[6,1],"3":[5,3],"4":[3,4],"5":[3,6],"6":[5,7],"7":[7,6]}
```

**4.2 Distance Calculation**

Write a function with two parameters named 'loc1' and 'loc2', representing the coordinates of two locations. The function should calculate and return the Euclidean distance between the two locations. Use this function to calculate the distance between the coordinates: (2, 1) and (5, 3).

Hint: The Euclidean distance between two locations (x1, y1) and (x2, y2) is

((x2-x1)^2+(y2-y1)^2)*0.5

```python
#Define a stop distance function.
def stop_distance(loc1, loc2):
    (x1, y1) = loc1
    (x2, y2) = loc2
    
#The given formula from the task.
    euc_distance = ((x2 - x1)**2 + (y2 - y1)**2)**0.5

    return euc_distance
```

```python
#Using the function to calculate the coordinates below.
stop_distance([2, 1], [5, 3])
```

**4.3 Nearest Stop**

Based on the dictionary defined in (4.1) and the function defined in (4.2). Write a program that asks the user to enter the coordinates of their location, uses a loop to find the bus stop closest to the user, and prints out the bus stop ID. If there are two stops that are equally close to the user, you only need to print one of them.

For example, if Stop 4 and Stop 5 are both the closest stops to the user, you can print Stop 4 or Stop 5.

```python
#Input values for the coordinates.
coordinate_x = float(input("Enter your coordinate (x): "))
coordinate_y = float(input("Enter your coordinate (y): "))

loc1 = [coordinate_x, coordinate_y]

#Closest stop has no value in this equation.
closest_stop = 0
shortest_distance = float("inf") #Using a float generates infinity positive values.

#For loop conditions for stop id, stop coordinations.
for stop_id, stop_coordinates in stop_information.items():
    distance = stop_distance(loc1, stop_coordinates)
    
#Condition for distance to find the shortest distance.
    if distance < shortest_distance:
        shortest_distance = distance
        closest_stop = stop_id

print(f"The closest stop is Stop {closest_stop}.")
```

**4.4 Estimated Travel Time**

Suppose the bus travels in one direction in the order of the stop IDs: 1, 2, ..., 7. Write a program that asks the user to enter the departure stop and destination stop. The program should calculate and print the estimated travel time based on the following information.

The result should be rounded to the first decimal place.

    - The bus travels at a speed of 2.5 minutes per kilometer on 
    normal routes (marked in green).
    - The bus travels at a speed of 4 minutes per kilometer on 
    busy routes (marked in red).

```python
#Create a dictionary for travel times between stops.
travel_times = {
    "1-2": 10.0,
    "2-3": 5.6,
    "3-4": 8.96,
    "4-5": 8.0,
    "5-6": 8.96,
    "6-7": 5.6
}

#Enter departure and destination stop values.
departure   = input("Enter departure stop: ")
destination = input("Enter destination stop: ")

#Validate that both stops exist.
if departure not in stop_information or destination not in stop_information:
    print("Invalid stop entered. Please enter bus stops 1-7.")
#Validate direction: the bus only travels from lower to higher stop IDs.
elif int(departure) >= int(destination):
    print("Departure stop must be before destination stop (bus travels in one direction: 1 → 7).")
else:
    #Calculate cumulative travel time.
    cum_route = 0
    stops = sorted(stop_information.keys(), key=int)
    stops = stops[stops.index(departure):stops.index(destination) + 1]

    #Give conditions for the stop value, which gives value to calculate the cumulative travel time.
    for i in range(len(stops) - 1):
        key = f"{stops[i]}-{stops[i + 1]}"
        cum_route += travel_times[key]

    #Define travel time.
    travel_time = round(cum_route, 1)

    print(f"Estimated travel time from Stop {departure} to Stop {destination} is {travel_time} minutes. Have a nice trip!")
```
